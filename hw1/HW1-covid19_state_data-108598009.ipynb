# Step 6 Use matplotlib show 2D images about data.
[Matplotlib](https://matplotlib.org/)
### Plot the distribution of two class (1. GDP < 58000, 2. GDP >= 58000) of COVID-19 State Data using different colors and different  marker where x-axis is the Pollution and y-axis the Mortality-rate.

```python
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame(covid, columns=['Pollution'])
df['Mortality_rate'] = covid.Deaths / covid.Infected
df['GDP'] = np.sign((covid.GDP-58000)/58000)

low_gdp = df.loc[df['GDP'] < 0]
high_gdp = df.loc[df['GDP'] >= 1]

fig, ax = plt.subplots()

ax.scatter(low_gdp.Pollution, low_gdp.Mortality_rate, s=50, c='blue', label='GDP < 58000')
ax.scatter(high_gdp.Pollution, high_gdp.Mortality_rate, s=50, c='red', label='GDP >= 58000')

ax.set_xlabel('Pollution', fontsize=15)
ax.set_ylabel('Mortality-rate', fontsize=15)
ax.legend()

ax.grid(True)

plt.show()
```

# Step 7
### Show the proportion of three class of COVID-19 State Data using pie chart.

**<font color = blue size=3> About the class: </font>** <p>
 1、 Mortality-rate < 0.02 <p>
 2、 Mortality-rate between 0.02 and 0.03 <p>
 3、Mortality-rate > 0.03 <p>

```python
labels = 'x < 0.02', '0.02 <= x <= 0.03', 'x > 0.03'
rate = df.Mortality_rate

sizes = [sum(rate < 0.02), sum(rate >= 0.02)-sum(rate > 0.03), sum(rate > 0.03)]

explode = (0, 0, 0.1) 

fig1, ax1 = plt.subplots()
sizes
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.set(aspect="equal", title='Mortality-rate x')

plt.show()
```
