### Homework Description
This assignment is to practice creating a bar graph grid using the tool such as MatPlotLib with Taipei’s traffic accident data from year 101 (2012) to 108 (2019). The files we provide are the data files (xxxTraffic accident.csv) of eight years and a correspondence table for mapping codes to certain fields (Traffic accident chart.csv). In the data file, there are fifteen attributes, including ***Year, Month, Day, Hour, Minute, Processing class, death toll, Injuries Person, Vehicle type, Sex, Age, Weather, Speed limit, Road pattern, Accident location***. With the data files and table, your homework should be done according to the following **requirements**:

#### There are six main topics to be shown in each year:  
   <font color=green> 1. Number of traffic accident death toll per month <br>
   2. Number of traffic accident injuries per month <br>
   3. Number of traffic accident injuries per vehicle type (take the top 10) <br>
   4. Number of traffic accident injuries per weather <br>
   5. Number of traffic accident injuries per road pattern (take the top 10) <br>
   6. Number of traffic accident injuries per accident location (take the top 10) <br>
   </font>
#### Each topic will be presented by a bar chart.   
#### For each year, use bar graph grid to show the six topics. Thus, there will be eight grids for years and each grid has six bar charts for each topic.  

#### On the other hand, please produce the grids for the six topics respectively. Each grid will have eight bar charts for the eight years.  

#### Please do not forget, in each chart, to have the topic on the top and label the x-axis and y-axis.
#### Sample:
 
**<p><p> <font color=green size=2> Year: 101 <font>** 
**<p><p> <font color=green size=2> Topic: top 10 Accident_location <font>** 

```python
import pandas as pd
import matplotlib.pyplot as plt

year = ['101','102','103','104','105','106','107','108']

def PlotByYear(year):
    
    file = pd.read_csv("./HW2_traffic_accident/" + str(year) + "Traffic_accident.csv")
    
    death_toll = file.groupby('Month')['death_toll'].sum().reset_index(name="death_toll").sort_values('death_toll', ascending=False)
    death_toll['Month'] = death_toll['Month'].apply(str)
    
    Month = file.groupby('Month')['Injuries_Person'].sum().reset_index(name="Injuries_Person").sort_values('Injuries_Person', ascending=False)
    Month['Month'] = Month['Month'].apply(str)
    
    Vehicle_type = file.groupby('Vehicle_type')['Injuries_Person'].sum().reset_index(name="Injuries_Person").sort_values('Injuries_Person', ascending=False).head(10)
    Vehicle_type['Vehicle_type'] = Vehicle_type['Vehicle_type'].apply(str)
    
    Weather = file.groupby('Weather')['Injuries_Person'].sum().reset_index(name="Injuries_Person").sort_values('Injuries_Person', ascending=False)
    Weather['Weather'] = Weather['Weather'].apply(str)
    
    Road_pattern = file.groupby('Road_pattern')['Injuries_Person'].sum().reset_index(name="Injuries_Person").sort_values('Injuries_Person', ascending=False).head(10)
    Road_pattern['Road_pattern'] = Road_pattern['Road_pattern'].apply(str)
    
    Accident_location = file.groupby('Accident_location')['Injuries_Person'].sum().reset_index(name="Injuries_Person").sort_values('Injuries_Person', ascending=False).head(10)
    Accident_location['Accident_location'] = Accident_location['Accident_location'].apply(str)
    
    fig, axes = plt.subplots(2,3, figsize = (24,10))
    plt.suptitle(('Year : ' + str(year)), fontsize=24)

    plt.subplot(2,3,1)
    plt.bar('Month', 'death_toll', data=death_toll)
    plt.title(str(year) + ' death_toll(Month)')
    plt.xlabel('Month')
    plt.ylabel('death_toll')

    plt.subplot(2,3,2)
    plt.bar('Month', 'Injuries_Person', data=Month)
    plt.title(str(year) +' Injuries_Person(Month)')
    plt.xlabel('Month')
    plt.ylabel('Injuries_Person')

    plt.subplot(2,3,3)
    plt.bar('Vehicle_type', 'Injuries_Person', data=Vehicle_type)
    plt.title(str(year) + ' Injuries_Person(TOP 10 Vehicle_type)')
    plt.xlabel('Vehicle_type')
    plt.ylabel('Injuries_Person')

    plt.subplot(2,3,4)
    plt.bar('Weather', 'Injuries_Person', data=Weather)
    plt.title(str(year) + ' Injuries_Person(TOP 10 Weather)')
    plt.xlabel('Weather')
    plt.ylabel('Injuries_Person')

    plt.subplot(2,3,5)
    plt.bar('Road_pattern', 'Injuries_Person', data=Road_pattern)
    plt.title(str(year) + ' Injuries_Person(TOP 10 Road_pattern)')
    plt.xlabel('Road_pattern')
    plt.ylabel('Injuries_Person')

    plt.subplot(2,3,6)
    plt.bar('Accident_location', 'Injuries_Person', data=Accident_location)
    plt.title(str(year) + ' Injuries_Person(TOP 10 Accident_location)')
    plt.xlabel('Accident_location')
    plt.ylabel('Injuries_Person')
```

```python
fig, axes = plt.subplots(2, 4, figsize = (20,10))
plt.suptitle('Topic : Number of traffic accident death toll per month', fontsize=24)
for i in range(101,109):
    file = pd.read_csv("./HW2_traffic_accident/" + str(i) + "Traffic_accident.csv")
    data = file.groupby('Month')['death_toll'].sum().reset_index(name="death_toll").sort_values('death_toll' , ascending=False)
    data['Month'] = data['Month'].apply(str)
    plt.subplot(2,4,(i-100))
    plt.bar('Month', 'death_toll', data = data)
    plt.title(str(i) + ' death_toll(Month)')
    plt.xlabel('Month')
    plt.ylabel('death_toll')
```
